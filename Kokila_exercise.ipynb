```python
#NAME PARSING EXERCISE

import os
import numpy
import pandas as pd
from nameparser import HumanName

filename="Kokila_Exercise1.xlsx"
writer=pd.ExcelWriter(r"C:\\Users\\Merit\\Desktop\\Data_standardization"+"\\"+filename,engine="xlsxwriter")

data_names = pd.read_excel('Kohila_Exercise.xlsx',sheet_name = 'Name Parsing')

namelist = []
for j in data_names["Name"].to_list():
    namelist.append(j)

title = []
first = []
last = []
middle = []
nick = []

for a in namelist:
    
    fullname = HumanName(a)
    
    #getting title
    if fullname.title == '':
        title.append("-")
    else:
        title.append(fullname.title)
    
    #getting first_name
    first.append(fullname.first)
    
    #getting last_name
    last.append(fullname.last)
    
    #getting middle_name
    if fullname.middle == '':
        middle.append("-")
    else:
        middle.append(fullname.middle)
    
    #getting nick_name
    if fullname.nickname == '':
        nick.append("-")
    else:
        nick.append(fullname.nickname)
        
names = zip(namelist,title,first,middle,last,nick)
namesdf = pd.DataFrame(names, columns = ['Full_Name','Title','First_name','Middle_name','Last_name','Nick_name'])
namesdf.to_excel(writer,sheet_name="Name Parsing",index=False)
```

```python
#ADDRESS PARSING
import os
import numpy
import pandas as pd
import usaddress

data = pd.read_excel('Kohila_Exercise.xlsx')
data
```

```python
address_list= []
for i in data["full_address"]:
    address_list.append(i)
    
#Getting US Address
us_address = []
for i in address_list:
    if "United States" in i:
        us_address.append(i)

address = []
for j in us_address:
    a = usaddress.tag(j, tag_mapping={
       'Recipient': 'address1',
       'AddressNumber': 'address1',
       'AddressNumberPrefix': 'address1',
       'AddressNumberSuffix': 'address1',
       'StreetName': 'address1',
       'StreetNamePreDirectional': 'address1',
       'StreetNamePreModifier': 'address1',
       'StreetNamePreType': 'address1',
       'StreetNamePostDirectional': 'address1',
       'StreetNamePostModifier': 'address1',
       'StreetNamePostType': 'address1',
       'LandmarkName': 'address1',
       'USPSBoxGroupID': 'address1',
       'USPSBoxGroupType': 'address1',
       'USPSBoxID': 'address1',
       'USPSBoxType': 'address1',
       'BuildingName': 'address1',
       'OccupancyType': 'address1',
       'OccupancyIdentifier': 'address1',
       'SubaddressIdentifier': 'address1',
       'SubaddressType': 'address1',
       'PlaceName': 'city',
       'StateName': 'zip_code',
       'ZipCode': 'zip_code',
        })
    address.append(a)
```

```python
country = []
zipcode = []
city = []
address1 = []

for i in address:
    #print(i[0])
    for key,value in i[0].items():
        if key == 'CountryName':
            country.append(value)
        elif key == 'zip_code':
            zipcode.append(value)
        elif key == 'city':
            city.append(value)
        elif key == 'address1':
            address1.append(value)
        else:
            address1.append('none')

address_data = zip(us_address,address1,city,zipcode,country)
address_data_df = pd.DataFrame(address_data,columns = ['Full_address','Address1','City','Zip_Code','Country'])
pd.set_option('display.max_rows',None)
address_data_df 

#Excel
address_data_df.to_excel(writer,sheet_name="Address Parsing",index=False)
writer.save()
writer.close()
```
